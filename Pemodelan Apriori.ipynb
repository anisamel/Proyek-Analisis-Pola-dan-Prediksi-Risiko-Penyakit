```python
import pandas as pd
df = pd.read_csv ("sampling_tahap1.csv")

df = df.astype(bool)

demografi_keywords = ["A", "B", "C", "D1","D2","D3","D4", "Q"]
pattern = '|'.join(demografi_keywords)
df = df.loc[:, ~df.columns.str.contains(pattern, case=False)]
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True,low_memory=True)
total_transactions = len(df)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets['frequency'] = (frequent_itemsets['support'] * total_transactions).astype(int)

jumlah_frequent_itemsets = len(frequent_itemsets)
print(f"Jumlah: {jumlah_frequent_itemsets}\n")

for i in range(1, frequent_itemsets['length'].max() + 1):
    print(f" {i} itemset")
    display(
        frequent_itemsets[frequent_itemsets['length'] == i]
        [['itemsets', 'support', 'frequency']]
        .sort_values(by='frequency', ascending=False)
    )
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules['support'] = (rules['support'] * 100).round(1)
rules['confidence'] = (rules['confidence'] * 100).round(1)
rules['lift'] = (rules['lift'] * 100).round(1)

jumlah_rules = len(rules)
print(f"Jumlah aturan: {jumlah_rules}")

rules.head(10)
```

```python
lifestyle_keywords = ['E', 'F', 'G', 'H1','H2','H3','H4', 'I', 'J1','J2','J3','J4', 'K', 'L', 'M', 'N', 'O1','O1','02','O3','O4' 'P']
lifestyle_factors = [col for col in df.columns if any(keyword in col for keyword in lifestyle_keywords)]

disease_factors = ['RD', 'RH', 'RJ', 'RO']
filtered_rules = rules[rules['antecedents'].apply(lambda x: all(item in lifestyle_factors for item in list (x)))]
filtered_rules = filtered_rules[filtered_rules['consequents'].apply(lambda x: all(item in disease_factors for item in list (x)))]

filtered_rules['antecedents'] = filtered_rules['antecedents'].apply(lambda x: ', '.join(list(x)))
filtered_rules['consequents'] = filtered_rules['consequents'].apply(lambda x: ', '.join(list(x)))

rules = filtered_rules.copy()
rules.drop(columns=['antecedent support', 'consequent support', 'leverage','conviction','zhangs_metric'], inplace=True)

jumlah_rules = len(rules)
print(f"Jumlah aturan: {jumlah_rules}")

rules.head(10)
```

```python
rules['antecedents'] = rules['antecedents'].astype(str)
rules['consequents'] = rules['consequents'].astype(str)

def filter_rules(rules, risiko_keyword, min_support, min_confidence, min_lift=1):
    filtered = rules[
        (rules['consequents'].str.contains(risiko_keyword)) &
        (rules['support'] >= min_support) &
        (rules['confidence'] >= min_confidence) &
        (rules['lift'] >= min_lift)
    ]
    return filtered

thresholds = {
    'RO': {'support': 2, 'confidence': 30},
    'RH': {'support': 3, 'confidence':30},
    'RD': {'support': 8, 'confidence': 30},
    'RJ': {'support': 9, 'confidence': 30}
}

for risiko, params in thresholds.items():
    filtered = filter_rules(rules, risiko, params['support'], params['confidence'])
    
    filename = f'rules_{risiko.lower()}.csv'
    filtered.to_csv(filename, index=False)
    print(f" {filename} berisi {filtered.shape[0]} rules.")
```
