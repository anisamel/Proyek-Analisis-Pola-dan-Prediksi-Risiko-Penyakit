```python
import pandas as pd
df = pd.read_csv ("polaMakan.csv")

#Statistik Atribut Numerik
desc = df.describe().round(0)
print("Statistik Deskriptif Numerik:\n")
print(desc)

#Distribusi Atribut Kategorikal
aktivitas_fisik = df['Aktivitas_Fisik'].value_counts(normalize=True) * 100
print("\nDistribusi Aktivitas Fisik (%):")
print(aktivitas_fisik.round(2))
IMT = df['IMT'].value_counts(normalize=True) * 100
print("\nDistribusi IMT (%):")
print(IMT.round(2))
tingkat_stres = df['Tingkat_Stres'].value_counts(normalize=True) * 100
print("\nDistribusi Tingkat Stres (%):")
print(tingkat_stres.round(2))
wlb = df['Work_Life_Balance'].value_counts(normalize=True) * 100
print("\nDistribusi Work Life Balance (%):")
print(wlb.round(2))

#Korelasi Variabel terkuat
df_numeric = df.select_dtypes(include=['int64', 'float64'])
corr = df_numeric.corr()
corr_matrix = df_numeric.corr().unstack().reset_index()
corr_matrix.columns = ['Var1', 'Var2', 'Correlation']

# Filter: nilai korelasi cukup kuat & bukan duplikat
strong_corr = corr_matrix[(abs(corr_matrix['Correlation']) >= 0.2) & (corr_matrix['Var1'] != corr_matrix['Var2'])]
strong_corr = strong_corr.drop_duplicates(subset=['Correlation']).sort_values(by='Correlation', ascending=False)
print(strong_corr)
```
