```python
import pandas as pd
df = pd.read_csv ("polaMakan.csv")
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
jumlah_duplikat = df.duplicated().sum()
print(f"Jumlah data duplikat: {jumlah_duplikat}")
```

```python
df_binned = df.copy()

df_binned['Usia_binned'] = pd.cut(df['Usia'], bins=[17, 25, 45, 59, 79], 
                                  labels=['Remaja', 'Dewasa', 'Paruh Baya', 'Lansia'])
df_binned['Frekuensi_Sayur_binned'] = pd.cut(df['Frekuensi_Sayur'], bins=[-1, 0, 2, 4, 6], 
                                             labels=['Sangat Jarang','Jarang', 'Sedang', 'Sering',])
df_binned['Frekuensi_Buah_binned'] = pd.cut(df['Frekuensi_Buah'], bins=[-1, 0, 2, 4, 6], 
                                            labels=['Sangat Jarang','Jarang', 'Sedang', 'Sering',])
df_binned['Frekuensi_Daging_Merah_binned'] = pd.cut(df['Frekuensi_Daging_Merah'], bins=[-1, 0, 2, 4, 6], 
                                                    labels=['Sangat Jarang','Jarang','Sedang','Sering',])
df_binned['Frekuensi_Makanan_Cepat_Saji_binned'] = pd.cut(df['Frekuensi_Makanan_Cepat_Saji'], bins=[-1, 0, 2, 4, 6], 
                                                          labels=['Sangat Jarang','Jarang','Sedang', 'Sering',])
df_binned['Frekuensi_Minuman_Bersoda_binned'] = pd.cut(df['Frekuensi_Minuman_Bersoda'], bins=[-1, 0, 2, 4, 6], 
                                                       labels=['Sangat Jarang','Jarang', 'Sedang', 'Sering',])
df_binned['Durasi_Tidur_binned'] = pd.cut(df['Durasi_Tidur'], bins=[0, 4, 6, 9], labels=['Sangat Pendek', 'Pendek', 'Normal'])

aktivitas_mapping = {
    'Tidak aktif': 'Tidak Aktif',
    '1-2 kali/minggu': 'Sedikit',
    '3-5 kali/minggu': 'Sedang',
    '>5 kali/minggu': 'Banyak',
}
df_binned['Aktivitas_Fisik_binned'] = df['Aktivitas_Fisik'].map(aktivitas_mapping)
```

```python
df_binned = pd.get_dummies(df_binned, columns=[
    'Usia_binned','Jenis_Kelamin', 'IMT', 
    'Riwayat_Penyakit_Keluarga','Frekuensi_Sayur_binned', 
    'Frekuensi_Buah_binned', 'Frekuensi_Daging_Merah_binned',
    'Frekuensi_Makanan_Cepat_Saji_binned','Frekuensi_Minuman_Bersoda_binned',
    'Frekuensi_Alkohol','Frekuensi_Kafein', 'Aktivitas_Fisik_binned', 'Durasi_Tidur_binned',
    'Tingkat_Stres', 'Kebiasaan_Merokok', 'Work_Life_Balance','Status_Ekonomi'], drop_first=False, dtype=int)
```

```python
columns_to_drop = [
    'Usia','Jenis_Kelamin', 'IMT', 'Riwayat_Penyakit_Keluarga', 'Frekuensi_Sayur', 
    'Frekuensi_Buah', 'Frekuensi_Daging_Merah', 'Frekuensi_Makanan_Cepat_Saji', 
    'Frekuensi_Minuman_Bersoda', 'Frekuensi_Alkohol', 'Frekuensi_Kafein', 
    'Aktivitas_Fisik', 'Durasi_Tidur', 'Tingkat_Stres', 'Kebiasaan_Merokok', 
    'Work_Life_Balance', 'Status_Ekonomi'
]
df_binned = df_binned.drop(columns=columns_to_drop, errors='ignore')
```
